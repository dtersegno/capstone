Using hashtags to assign a game type
---

```python
import json
import pandas as pd
import numpy as np
```

```python
hashtags = pd.read_csv('./all_hashtags.csv')['hashtag'].value_counts()
```

```python
hashtags
```

```python
#list most popular hashtags
list(hashtags.index)
```

```python
# a few of these are distinctively from a certain kind of game.
# make a dict that converts these.
hashtag_to_game = {
    'ポケモンWordle':'JPPokemon',
    'Wordle':'Wordle',
    'wordle':'Wordle',
    'WordleCAT':'Catalan',
    'вордли':'Russian',
    'Heardle':'Heardle',
    '워들':'Korean',
    '한글':'Korean',
    'Korean':'Korean',
    'wordleTR':'Turkish',
    'AlWird':'Arabic',
     'galego':'Galego',
     'WordleGalego':'Galego',
    'DailyWordleClub':'WordleNYT',
    'WordleEnglish':'WordleNYT',
    'WordleEN':'WordleNYT',
    'WordleEUS':'Basque',
    'hadesjSlova':'Czech',
    'WORDLE_TAMIL':'Tamil',
    'WizardingWordle':'Wizard',
    'Enstarsdle':'Enstars',
    'WordleEspañol':'Spanish',
    'wordleES':'Spanish',
    '嘘wordle':'Japanese',
    '百人一首Wordle':'Japanese',
    'Quordle':'Quordle',
    'ことのはたんご':'kotonoha-tango',
    'NYTimes':'WordleNYT'
}
```

```python
#export dict
with open('./hashtag_to_game.json','w') as file:
    file.write(json.dumps(hashtag_to_game))
```

```python
#check it worked
with open('./hashtag_to_game.json','r') as file:
    loaded_hashtags = json.loads(file.read())
loaded_hashtags
```
